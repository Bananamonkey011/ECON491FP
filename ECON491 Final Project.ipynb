```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import geopandas as gpd # geospatial data processing
import numpy as np # linear algebra operations
import matplotlib.pyplot as plt # plotting
import sklearn.linear_model as lm # linear regression
from sklearn.feature_selection import SequentialFeatureSelector # subset selection
from sklearn.model_selection import train_test_split # train-test split of data
from sklearn.metrics import mean_squared_error # MSE calculator
```

# Import and Clean Data

### Importing the Data

```python
housing_org = pd.read_csv("housing.csv", low_memory=False)  # read in data
crimes_org = pd.read_csv("crimes.csv", low_memory=False)
parks_org = gpd.read_file("park.geojson")
tract_boundaries = gpd.read_file('b.geojson')
# shoudl take around 30-40 seconds to run
print(housing_org.shape)
```

### Filtering the Data

```python
crimes_filt = crimes_org[crimes_org['Latitude'] > 41 ]
crimes_filt = crimes_filt[crimes_filt['Latitude'] < 43]
crimes_filt = crimes_filt[crimes_filt['Longitude'] > -88]
crimes_filt = crimes_filt[crimes_filt['Longitude'] < -87.5]

parks_org.x_coord = parks_org.x_coord.astype(float)
parks_org.y_coord = parks_org.y_coord.astype(float)
parks_org.objectid_1 = parks_org.objectid_1.astype(int)
parks_org.gisobjid = parks_org.gisobjid.astype(float)
parks_org.park_no = parks_org.park_no.astype(float)

parks_filt = parks_org[(parks_org['y_coord']) > 41 ]
parks_filt = parks_filt[parks_filt['y_coord'] < 43]
parks_filt = parks_filt[parks_filt['x_coord'] > -88]
parks_filt = parks_filt[parks_filt['x_coord'] < -87.5]
```

```python
housing_filt = housing_org[
    [
        "Property Index Number",
        "Property Class",
        "Township Name",
        "Number of Active 288s (HIEs)",
        "Number of Ended 288s (HIEs)",
        "Sale Price",
        "Land Square Feet",
        "Large Home Indicator",
        "Age",
        "Apartments",
        "Wall Material",
        "Roof Material",
        "Rooms",
        "Bedrooms",
        "Basement",
        "Basement Finish",
        "Central Heating",
        "Central Air Conditioning",
        "Fireplaces",
        "Attic Type",
        "Full Baths",
        "Half Baths",
        "Cathedral Ceiling",
        "Garage 1 Size",
        "Garage 1 Material",
        "Garage 1 Attached",
        "Garage 1 Area",
        "Building Square Feet",
        "Repair Condition",
        "Attic Finish",
        "Renovation",
        "Porch",
        "Tract GEOID",
        "O'Hare Noise Indicator",
        "FEMA Floodplain",
        "Flood Risk Factor",
        "Flood Risk Direction",
        "Road Proximity < 100 Feet",
        "Road Proximity 101 - 300 Feet",
        "Elementary/Middle School District",
        "High School District",
        "Longitude",
        "Latitude",
        "Tax Rate",
        "Tract Median Income",
        "Modeled Sale Year",
        "Modeled Sale Month",
        "Modeled Sale Day",
        "Modeled Sale During School Year",
        "Modeled Sale During Holidays",
    ]
]
print(housing_filt.shape)
housing_filt.dropna(subset=["Sale Price"], inplace=True)
print(housing_filt.shape)
```

### Cleaning Data

```python
if 'Property Index Number' in housing_filt.columns: # set PIN as index
	housing_filt.set_index('Property Index Number', inplace=True)
print(housing_filt.shape)
for x in housing_filt.columns: # convert all columns to numeric 
	if housing_filt[x].dtype == 'object':
		housing_filt[x].replace(housing_filt[x].unique(), range(len(housing_filt[x].unique())), inplace=True)
		housing_filt[x] = housing_filt[x].astype('category')

print(housing_filt.shape)
housing_filt = housing_filt.dropna()
print(housing_filt.shape)
```

### Constructing Geospatial DataSets

```python
# tract_boundaries.geoid10 = tract_boundaries.geoid10.astype(int)

housing = gpd.GeoDataFrame(housing_filt, crs="EPSG:4326", geometry=gpd.points_from_xy(housing_filt.Longitude, housing_filt.Latitude))
housing.rename(columns={'Tract GEOID': 'geoid10'}, inplace=True)
print(housing.shape)

crimes = gpd.GeoDataFrame(
    crimes_filt,
    crs="EPSG:4326",
    geometry=gpd.points_from_xy(crimes_filt. Longitude, crimes_filt.Latitude),
)
crimes = gpd.sjoin(crimes, tract_boundaries, how="left", predicate="within")

parks = gpd.GeoDataFrame(
    parks_filt,
    crs="EPSG:4326",
    geometry=gpd.points_from_xy(parks_filt.x_coord, parks_filt.y_coord),
)
parks = gpd.sjoin(parks, tract_boundaries, how="left", predicate="within")
```

### Create dataset by districts

```python
crime_counts = crimes.groupby("geoid10", as_index=False)["ID"].count()
park_counts = parks.groupby("geoid10", as_index=False)["objectid_1"].count()
housing_avgs = (
    housing.groupby("geoid10", as_index=False)["Sale Price"].mean().reset_index()
)

housing_avgs.geoid10 = housing_avgs.geoid10.astype(int)

districts = crime_counts.merge(park_counts, on="geoid10")
districts.geoid10 = districts.geoid10.astype(int)
districts = districts.merge(housing_avgs, on="geoid10")
districts.rename(columns={"Sale Price": "avg_sale"}, inplace=True)
districts.rename(columns={"ID": "crime_count"}, inplace=True)
districts.rename(columns={"objectid_1": "park_count"}, inplace=True)
districts = districts.fillna(0)
```

### Add district data to individual house data

```python
print(housing.shape)
housing = housing.merge(districts, on="geoid10", how="left")
print(housing.shape)
```

# Impact of Individual House Price

### Train Test Split

```python
housing2 = housing.drop('geometry', axis = 1)
housing2 = housing2.drop('avg_sale', axis = 1)
housing2 = housing2.dropna()
X = housing2.drop('Sale Price', axis = 1)
y = housing2['Sale Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=491)
```

### Subset Selection + Linear Regression

For Linear Regression with subsect selection, we see that we have a Test MSE of $44440433153.70373. We can break this down further to its individual error of $210809.00. This show us that our linear model, on average, is off by about $200k. We also see by the feature importance graph that the model is selecting the location (via long. and lat.) as the most important variables. 

Large home indicator is also a significantly important variable. This falls in line with common reasoning. The size of a home would indicate how much the house would sell for almost as much as the location of the house. We also see that renovation is a key indicator of the model as well, indicating that much of the dataset with significantly different home prices might be those that have recently had renovations done to them.

```python
model = lm.LinearRegression(fit_intercept=True)
sfs = SequentialFeatureSelector(model, n_features_to_select=16, direction='forward')

sfs.fit(X_train, y_train)
x_train_new = sfs.transform(X_train)
lmodel = model.fit(x_train_new, y_train)
y_pred = model.predict(sfs.transform(X_test))
mse = mean_squared_error(y_test, y_pred)
print(mse)
print(np.sqrt(mse))
```

```python
importance = np.abs(lmodel.coef_)
features = sfs.get_feature_names_out()
plt.bar(height=importance, x=features)
plt.title("Feature importances from Subset Selection")
plt.xticks(rotation=60, ha = 'right')
plt.show()
```

### Ridge

Ridge Regression shows slightly better Test MSE and indiviual error at $43186778412.53 and $207814.288 per house. Longitude, Large Home Indicator, Latitude, Renovation, & Road Proximity being less than 100 Feet where similar to subset regression, the variables that affected the model most. We even see negative values being associated with the road proximity like one might expect.

```python
lm2 = lm.Ridge(fit_intercept=True)
lm2.fit(X_train, y_train)
y_pred = lm2.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(mse)
print(np.sqrt(mse))

feature_importances = lm2.coef_
feature_importances_dict = {col: coef for coef, col in zip(lm2.coef_, X_train.columns)}
sorted_importances = sorted(feature_importances_dict.items(), key=lambda x: abs(x[1]), reverse=True)
plt.figure(figsize=(10, 6))
plt.bar([x[0] for x in sorted_importances], [x[1] for x in sorted_importances])
plt.xlabel('Feature Name')
plt.ylabel('Coefficient')
plt.title('Ridge Feature Importances')
plt.xticks(rotation=90)
plt.show()
```

### Lasso

Lasso performs significantly worse than both Ridge and Subset Selection, with Test MSE and indiviual error at $67632905016.86 and $260063.27 per house. Almost identically to Ridge, we see that: Longitude, Large Home Indicator, Latitude, Renovation, & Road Proximity being less than 100 Feet were the variables that affected the model most, we also see negative values being associated with the road proximity.

```python
lm2 = lm.LassoCV(fit_intercept=True)
lm2.fit(X_train, y_train)
y_pred = lm2.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(mse)
print(np.sqrt(mse))

feature_importances = lm2.coef_
feature_importances_dict = {col: coef for coef, col in zip(lm2.coef_, X_train.columns)}
sorted_importances = sorted(feature_importances_dict.items(), key=lambda x: abs(x[1]), reverse=True)
plt.figure(figsize=(10, 6))
plt.bar([x[0] for x in sorted_importances], [x[1] for x in sorted_importances])
plt.xlabel('Feature Name')
plt.ylabel('Coefficient')
plt.title('Lasso Feature Importances')
plt.xticks(rotation=90)
plt.show()
```

# Impact of neighborhood

We analyzed crime & park data in Chicago to see if there was any correlation as well. This is because:

- Crime can be a major factor affecting the price of homes. One of the main reasons is safety concerns. Potential buyers generally prefer to live in a safe neighborhood, and high crime rates can make an area feel unsafe. This can discourage people from buying homes in that neighborhood, leading to lower demand and, ultimately, lower prices.
 
- In addition to safety concerns, crime can also increase insurance costs. Home insurance companies typically charge higher premiums for properties located in high-crime areas. This can make it more expensive for homeowners to insure their properties, which can reduce their overall affordability and desirability.
 
- Crime can also result in property damage, such as broken windows or graffiti. This can lower the curb appeal of a home and reduce its value. The cost of repairing this damage can also be a factor that negatively affects home prices.
 
- High crime rates can create a negative perception of a neighborhood. Even if the crime rate is not particularly high, a negative perception of the area can still impact the price of homes. This negative perception can reduce demand and lower home prices

The number of parks in a neighborhood can be an important factor affecting the price of homes in the area. Having access to green spaces, such as parks, is desirable for many homebuyers, as they provide opportunities for recreation, relaxation, and social interaction. Thus, the availability and proximity of parks can contribute positively to the value of homes in the surrounding area.

Generally, homes located in neighborhoods with a higher number of parks tend to have higher property values than those in areas with fewer parks. This is because the presence of parks can enhance the quality of life for residents, which can increase demand for homes in the area. Additionally, proximity to parks can be a desirable feature for many homebuyers, particularly those with children or pets, which can also drive up home prices.

However, the relationship between the number of parks and home prices can vary depending on the location and type of parks. For example, parks that are well-maintained, have a variety of amenities, and are located within a reasonable distance of homes may have a greater positive impact on home prices than parks that are poorly maintained or located far away. Similarly, the presence of certain types of parks, such as those with sports facilities or walking trails, may be more desirable for some homebuyers than others, which can also affect home prices.

We can see from this last plot, the distribution of average home prices in an area based off how much crime there is -- as well as how many park district facilities there are. This data isn't perfect -- since there is not a 1:1 relationship between number of parks and number of park district facilities, however, it should give us an idea of the effect.

We see, like one might expect, that there is a significantly lower home prices in areas where the crime rate is much higher. The concentration of the prices is right at the bottom of the average house price -- unlike in areas where crime is much lower where we see a more typical distribution of home prices.

Somewhat more interestingly we see that as park count increases, the house price does not actually increase. 

This could be because, the quality of the parks may be poor, which can offset any potential positive effects on home values. If the parks are poorly maintained or do not have desirable features, such as sports facilities or walking trails, they may not be attractive to homebuyers and may not contribute positively to home values.

Secondly, the location of the parks may not be ideal for homebuyers. If the parks are located far away from residential areas or in undesirable locations, such as near industrial sites or highways, they may not have a positive impact on home values.

Other factors may be more important in determining home values in the area. Like we have tested, access to good schools, proximity to public transportation, and low crime rates may be more significant factors in driving up home prices than the number of parks in the area.

# Conclusion

In our research, we compared three machine learning models - Linear Regression with subset selection, Ridge Regression, and Lasso Regression - to predict housing prices in Chicago. Each model incorporated a wide range of variables, including crime rates, park proximity, and other factors influencing home values. The Ridge Regression model provided the best performance with a Test MSE of $43186778412.53 and an individual error of $207814.288 per house. In contrast, Linear Regression with subset selection had a Test MSE of $44440433153.70373 and an individual error of $210809.00, while Lasso Regression had a Test MSE of $67632905016.86 and an individual error of $260063.27 per house.

The relatively high errors observed in all three models could be attributed to several factors. Firstly, the large number of variables included in the models might have introduced multicollinearity or increased complexity, making it difficult for the models to discern the true underlying relationships between the variables and housing prices. Secondly, there may be potential influential outliers or noise in the data, which can adversely affect the model's predictive accuracy. Lastly, the models we used might not be able to fully capture the complex, nonlinear relationships between housing prices and the predictors, resulting in high errors.

To improve the performance of our models, we could consider exploring other machine learning techniques that can better handle complex relationships and high-dimensional data. Some potential models include:

Decision Trees and Random Forests: These models can handle nonlinear relationships and interactions between variables, which might help to capture the complex patterns in the data. Random Forests, in particular, can provide a more robust and accurate model by aggregating the predictions of multiple Decision Trees.

Gradient Boosting Machines (GBM): GBMs are a powerful ensemble technique that builds a strong model by iteratively adding weak models, usually decision trees, to the ensemble. GBMs can handle complex interactions and nonlinearities in the data and often provide better predictive accuracy compared to single models.

Support Vector Machines (SVM) with a nonlinear kernel: SVMs can be used for regression tasks, and by using a nonlinear kernel, such as the Radial Basis Function (RBF) kernel, they can effectively capture complex, nonlinear relationships in the data.

Neural Networks: Deep learning techniques, such as neural networks, can model highly complex relationships between variables and are especially suited for handling large and high-dimensional datasets. However, they require careful tuning and can be computationally expensive.

Before applying these alternative models, it is essential to address any data quality issues, such as outliers or noise, that might be contributing to the high errors. A major failing of our research and project was to produce a dataset of consistent and reliable data -- our work was significantly hampered by constant issues with the data not fitting proprly, or having NA values for many of the columns we had.

Additionally, feature engineering techniques, such as dimensionality reduction or feature selection, can be employed to reduce the number of variables in the models and potentially enhance their performance. By exploring these alternative models and refining our dataset, we could potentially achieve more accurate predictions of housing prices in Chicago.

Furthermore, our exploration of crime rates and park proximity unveiled some interesting insights. As expected, areas with higher crime rates exhibited significantly lower home prices, demonstrating the impact of safety concerns on housing values. However, the number of parks in a neighborhood did not have a clear positive relationship with home prices. This could be attributed to various factors, such as park quality, location, or other more significant determinants of home values like access to good schools, proximity to public transportation, and low crime rates.

By incorporating these factors into our machine learning model, we were able to gain a more comprehensive understanding of housing values in Chicago. Our study also responded to the call for home valuation methodologies that move beyond racially biased appraisal practices and consider actual determinants of housing quality and safety. The findings of this research not only contribute to the existing literature on housing price prediction but also serve as a valuable resource for policymakers, real estate professionals, and homeowners in understanding the complex interplay of factors that shape housing values in urban areas.
